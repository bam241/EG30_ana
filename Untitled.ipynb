```python
from cymetric import timeseries as com
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cymetric as cym
from cymetric import graphs

```

```python
db_mlp = cym.dbopen('mlp_stoping_blk/cyclus.sqlite')
ev_mlp = cym.Evaluator(db=db_mlp, write=False)
```

```python
db_mlp_decay = cym.dbopen('mlp_stoping_blk_decay/cyclus.sqlite')
ev_mlp_decay = cym.Evaluator(db=db_mlp_decay, write=False)
```

```python
db_recipe = cym.dbopen('recipe_up/cyclus.sqlite')
ev_recipe = cym.Evaluator(db=db_recipe, write=False)
```

```python
def month2year(df, mode=0, division=12):
    dfn = pd.DataFrame(columns=['Time','Mass'])
    val = 0
    for index, row in df.iterrows():
        if mode == 0:
            val = row['Mass']
        else:
            val += row['Mass']
        if row['Time'] % division == 0:
            if mode == 2:
                val *= 1./double(division)
            dfn.loc[int(row['Time'] / division)] = int(row['Time'] / 12)
            dfn.loc[int(row['Time'] / division)]['Mass'] = val
            val = 0
    return dfn
            
```
